```python

import psycopg2

# Conectar con PostgreSQL desde Spark
conn = psycopg2.connect(
    dbname="psqldbSpark",
    user="cachoza_user",
    password="cachoza_password",
    host="postgresdb_cachoza",  # Nombre del servicio en Docker
    port=5432
)

# Crear tabla
query_create = """
CREATE TABLE IF NOT EXISTS usuarios (
    id SERIAL PRIMARY KEY,
    nombre VARCHAR(100) NOT NULL,
    email VARCHAR(255) UNIQUE NOT NULL,
    fecha_registro TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
"""

# Ejecutar la consulta
cursor = conn.cursor()
cursor.execute(query_create)
conn.commit()
cursor.close()
conn.close()

print(" Tabla 'usuarios' creada correctamente desde Spark.")




#from pyspark.sql import SparkSession
#
# Iniciar sesión de Spark
#spark = SparkSession.builder \
#    .appName("PostgreSQL_Spark") \
#    .config("spark.jars", "/path/to/postgresql-<version>.jar") \
#    .getOrCreate()
#
# Datos de usuario
#data = [("Miguel", "miguel@example.com")]
#columns = ["nombre", "email"]
#
# Crear DataFrame de Spark
#usuarios_df = spark.createDataFrame(data, columns)
#
# Propiedades de la conexión
#jdbc_url = "jdbc:postgresql://postgresdb_cachoza:5432/psqldbSpark"
#db_properties = {
#    "user": "cachoza_user",
#    "password": "cachoza_password",
#    "driver": "org.postgresql.Driver"
#}
# Escribir datos en la tabla
#usuarios_df.write.jdbc(url=jdbc_url, table="usuarios", mode="append", properties=db_properties)
#print(" Datos insertados correctamente en la tabla 'usuarios'.")
#df = spark.read.jdbc(url=jdbc_url, table="usuarios", properties=db_properties)
#df.show()
```
