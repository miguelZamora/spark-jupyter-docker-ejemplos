```python
from pyspark.sql import SparkSession

# Iniciar una sesión Spark
spark = SparkSession.builder \
    .appName("Ejecutor Simple") \
    .config("spark.executor.instances", "1") \
    .config("spark.executor.memory", "2g") \
    .config("spark.executor.cores", "2") \
    .getOrCreate()

# Verificar la configuración de los ejecutores
print(spark.conf.get("spark.executor.instances"))
print(spark.conf.get("spark.executor.memory"))
print(spark.conf.get("spark.executor.cores"))

# Crear un RDD básico
rdd = spark.sparkContext.parallelize([1, 2, 3, 4, 5])
print(rdd.collect())

# Detener la sesión
spark.stop()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Jupyter-Spark Integration") \
    .master("spark://spark-master:7077") \
    .config("spark.driver.memory", "2g") \
    .config("spark.executor.memory", "2g") \
    .getOrCreate()

print(spark)

data = [(1, "Chile"), (2, "Argentina"), (3, "Brasil")]
df = spark.createDataFrame(data, ["ID", "País"])
df.show()



```

```python
print(spark.sparkContext.defaultParallelism)  # Número de núcleos disponibles
print(spark.sparkContext.appName)  # Nombre de la aplicación
print(spark.sparkContext.master)  # Dirección del master
```

```python
from pyspark.sql import Row

datos = [Row(id=1, nombre="Miguel"), Row(id=2, nombre="Gomi"), Row(id=3, nombre="Python")]
df = spark.createDataFrame(datos)

df.createOrReplaceTempView("usuarios")

spark.sql("SELECT * FROM usuarios WHERE id > 1").show()
```

```python
from pyspark.sql import SparkSession
import os

# Configura la variable de entorno ANTES de crear la sesión (opcional)
os.environ['PYSPARK_SUBMIT_ARGS'] = '--jars /path/to/postgresql-42.6.0.jar pyspark-shell'

# Crea la sesión de Spark con la configuración del driver
spark = SparkSession.builder \
    .appName("PostgreSQL_Spark") \
    .config("spark.jars", "/path/to/postgresql-42.6.0.jar") \
    .getOrCreate()

# Datos de conexión
jdbc_url = "jdbc:postgresql://127.0.0.1:5432/psqldbSpark"
db_properties = {
    "user": "cachoza_user",
    "password": "cachoza_password",
    "driver": "org.postgresql.Driver"
}

try:
    # Cargar tabla de usuarios (asegúrate que el nombre de la tabla es correcto)
    usuarios_df = spark.read.jdbc(
        url=jdbc_url, 
        table="usuarios",  # Verifica si es "usuario" o "usuarios"
        properties=db_properties
    )
    
    # Mostrar datos
    usuarios_df.show()
except Exception as e:
    print("Error detallado:")
    print(e)
    print("\nMensaje completo:")
    e.java_exception.printStackTrace()
```
