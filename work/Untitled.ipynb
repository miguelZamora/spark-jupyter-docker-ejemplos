```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Postgres JDBC") \
    .config("spark.jars", "/opt/spark/jars/postgresql-42.7.7.jar") \
    .getOrCreate()
```

```python
jdbc_url = "jdbc:postgresql://postgresdb_cachoza:5432/psqldbSpark"
db_properties = {
    "user": "cachoza_user",
    "password": "cachoza_password",
    "driver": "org.postgresql.Driver"
}

df = spark.read.jdbc(url=jdbc_url, table="usuarios", properties=db_properties)
df.show()
```
