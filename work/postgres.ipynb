```python
import psycopg2
```

```python


# Configurar los parámetros de conexión
conn = psycopg2.connect(
    host="postgresdb_cachoza",  # Puede ser 'localhost' o el nombre del contenedor en Docker
    database="psqldbSpark",
    user="cachoza_user",
    password="cachoza_password",
    port="5432"  # Puerto por defecto de PostgreSQL
)
```

```python
# conectar primero y luego una consulta simple de la version de postgres instalada by Docker.
```

```python
# Crear un cursor
cursor = conn.cursor()

# Ejecutar una consulta de prueba
cursor.execute("SELECT version();")
print(cursor.fetchone())

# Cerrar la conexión
cursor.close()
conn.close()

```

```python
from pyspark.sql import SparkSession
import os

# Configura la variable de entorno ANTES de crear la sesión (opcional)
os.environ['PYSPARK_SUBMIT_ARGS'] = '--jars ./jars/postgresql-42.5.4.jar pyspark-shell'

# Crea la sesión de Spark con la configuración del driver
spark = SparkSession.builder \
    .appName("PostgreSQL Connection") \
    .config("spark.driver.extraClassPath", "/opt/spark/jars/postgresql-42.3.9.jar") \
    .getOrCreate()


# Datos de conexión
jdbc_url = "jdbc:postgresql://127.0.0.1:5432/psqldbSpark"
db_properties = {
    "user": "cachoza_user",
    "password": "cachoza_password",
    "driver": "org.postgresql.Driver"
}

try:
    # Cargar tabla de usuarios (asegúrate que el nombre de la tabla es correcto)
    usuarios_df = spark.read.jdbc(
        url=jdbc_url, 
        table="usuario",  # Verifica si es "usuario" o "usuarios"
        properties=db_properties
    )
    
    # Mostrar datos
    usuarios_df.show()
except Exception as e:
    print("Error detallado:")
    print(e)
    print("\nMensaje completo:")
    e.java_exception.printStackTrace()
```
