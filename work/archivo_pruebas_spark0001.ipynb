```python
from pyspark.sql import SparkSession

# Crear sesión de Spark
spark = SparkSession.builder \
    .appName("MiSparkSession") \
    .master("spark://spark-master:7077") \
    .getOrCreate()

# Verificar la sesión
print(spark.version)
```

```python
data = [(1, "Miguel"), (2, "Gomi"), (3, "SparkMaster")]
df = spark.createDataFrame(data, ["id", "nombre"])

# Mostrar los primeros registros
df.show()
```

```python
df.createOrReplaceTempView("usuarios")

spark.sql("SELECT * FROM usuarios WHERE id = 2").show()
```
